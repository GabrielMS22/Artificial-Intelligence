# Arbol genealogico con Python


```python
from pyswip import Prolog, registerForeign
from pyswip import Functor, Variable, Query,call
#Gabriel Melchor Sanchez
#Arbol genealogico utilizando Prolog, el motor es Python

def Arbol():
    prolog = Prolog()
    father = Functor("father", 2)
    mother = Functor("mother", 2)
    assertz = Functor("assertz",1)
    X = Variable()
    Y = Variable()
    #vis es la variable con la que activamos la consulta cuando ya se tienen datos agregados
    vis=0
    #Ciclo controlador para agregar datos al arbol
    print("Creacion de Arbol genealogico")
    while True:
        print("\n")
        print("1-Agregar")
        if vis==1:
            print("2-Verificar Arbol genealogico")
        print("3-Salir")
        r=int(input())
        #Si r es igual a 1, se agregaran hijos
        if r == 1:
            add = 1
            
            print("\nAgregar al Arbol")
            while add == 1:
                print("\n1-Agregar Hijos")
                print("Cualquier otro numero para Salir")
                add=int(input())
                if add==1:
                    print("Dame el nombre del hijo")
                    hijo = input()
                    print("Ahora dame el nombre de su padre")
                    padre = input()
                    print("Tambien dame el nombre de su madre")
                    madre = input()
                    #Agregamos con assertz los respectivos padres del hijo
                    prolog.assertz("father("+padre+","+hijo+")")
                    prolog.assertz("mother("+madre+","+hijo+")")
                    #Variable que indica que se han agregado personas al arbol
                    #por ende la opcion de verificar Arbol se vuelve disponible
                    vis=1
        #Si r es igual a 2, se verificara el arbol con los metodos disponibles
        
        if r == 2:
            listing = Functor("listing", 1)
            call(listing(father))
            #Desplegamos opciones para verificar los agregados en el arbol
            print("\nSelecciona la opcion que quieras investigar del Arbol Genealogico")
            print("Notese que si el arbol no cuenta con la informacion necesaria, se notificara el error")
            print("1-Dame el padre de...")
            print("2-Dame la madre de...")
            print("3-Dame el hijo  de...")
            print("4-Dame el hermano de...")
            cons = int(input())
            

            if cons == 1:
                print("\nEscriba el nombre de la persona para buscar a su padre")
                op=input()
                
                for padre in prolog.query("father(Y,"+op+")"):
                    print(padre["Y"],"es padre de",op)
                
            if cons == 2:
                print("\nEscriba el nombre de la persona para buscar a su madre")
                op=input()
                
                for madre in prolog.query("mother(Y,"+op+")"):
                    print(madre["Y"],"es madre de",op)
            if cons == 3:
                 
                    print("1-Madre\n2-Padre ")
                    #Variable que indica si se trata de un Padre o una Madre
                    quien=int(input())
                    if quien==1:
                        print("\nEscriba el nombre de la persona para buscar a su hijos")
                        op=input()
                        for hijo in prolog.query("mother("+op+",Y)"):
                            
                            print(hijo["Y"],"es hijo de",op)
                    if quien==2:
                        print("\nEscriba el nombre de la persona para buscar a su hijos")
                        op=input()
                        for hijo in prolog.query("father("+op+",Y)"):
                            
                            print(hijo["Y"],"es hijo de",op)
            if cons == 4:
                print("\nEscribe la persona de la que deseas conocer sus hermanos")
                herm=input()
                for hermano in prolog.query("father(Y,"+herm+"),father(Y,B),B\="+herm):
                    print(hermano["B"],"es hermano de",herm)
                      
                            
                            
                
                    
        #Si r es 3, se acaba la ejecucion
        if r == 3:
            print("Hasta luego")
            break
Arbol()





    
```
